```python
# import libraries
import pandas as pd
import numpy as np

import re, nltk

# Sklearn
from sklearn.decomposition import LatentDirichletAllocation, TruncatedSVD
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import GridSearchCV

# nltk
import nltk
nltk.download('wordnet')

from nltk.stem.wordnet import WordNetLemmatizer
```

```python
# read in the data with labels_df
labels_df = pd.read_csv("insta_data_with_labels.csv")
```

```python
labels_df.head()
```

```python
labels = labels_df.drop(labels_df.columns[[0,1,2,3,4,5]], axis = 1)
```

```python
labels.head()
```

```python
labels.shape
```

```python
labels = labels.dropna()
```

```python
labels.head()
```

```python
# lemmatizer
def ltizer(x):
    lemma = WordNetLemmatizer() 
    return lemma.lemmatize(x)
```

```python
# conver to series
labels["labels"] = labels.iloc[:,0].apply(ltizer)
```

```python
# create word document matrix
labels_np = labels.values
```
